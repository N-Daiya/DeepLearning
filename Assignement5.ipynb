## DATA AUGUMENTATION -FLOW FROM DATAFRAME

```
from bs4 import BeautifulSoup
import numpy as np
import requests
import cv2
import PIL.Image
import urllib

flower_page = requests.get("http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n11669921")#Flowers
print(flower_page.content)
flower_soup = BeautifulSoup(flower_page.content, 'html.parser')#puts the content of the website into the soup variable, each url on a different line




veg_page = requests.get("http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n07707451")#Vegetables
print(veg_page.content)

from bs4 import BeautifulSoup
veg_soup = BeautifulSoup(veg_page.content, 'html.parser')#puts the content of the website into the soup variable, each url on a different line


```

```

#code part 2
flower_str_soup=str(flower_soup)#convert soup to string so it can be split
type(flower_str_soup)
flower_split_urls=flower_str_soup.split('\r\n')#split so each url is a different possition on a list
print(len(flower_split_urls))#print the length of the list so you know how many urls you have

#code part 2.2
veg_str_soup=str(veg_soup)#convert soup to string so it can be split
type(veg_str_soup)
veg_split_urls=veg_str_soup.split('\r\n')#split so each url is a different possition on a list
print(len(veg_split_urls))


!mkdir /content/fv_train/ 
!mkdir /content/fv_test/ 

img_rows, img_cols = 32, 32 #number of rows and columns to convert the images to
input_shape = (img_rows, img_cols, 3)#format to store the images (rows, columns,channels) called channels last

def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urllib.request.urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image

n_of_training_images=150#the number of training images to use
for progress in range(n_of_training_images):#store all the images on a directory
 
    if(progress%20==0):
        print(progress)
    if not flower_split_urls[progress] == None:
      try:
        I = url_to_image(flower_split_urls[progress])
        if (len(I.shape))==3: #check if the image has width, length and channels
          save_path = '/content/fv_train/afr.'+str(progress)+'.jpg'#create a name of each image
          cv2.imwrite(save_path,I)

      except:
        None

for progress in range(n_of_training_images):#store all the images on a directory
  
    if(progress%20==0):
        print(progress)
    if not veg_split_urls[progress] == None:
      try:
        I = url_to_image(veg_split_urls[progress])
        if (len(I.shape))==3: #check if the image has width, length and channels
          save_path = '/content/fv_train/ind.'+str(progress)+'.jpg'#create a name of each image
          cv2.imwrite(save_path,I)

      except:
        None
```

##Save images to the folder


```
# Path to the train folder
import os
original_train = '/content/fv_train/'
 
filenames = os.listdir(original_train)
categories = []
for filename in filenames:
    category = filename.split('.')[0]
    if category == 'afr':
        categories.append('0')
    else:
        categories.append('1')
```

## Create dataframe from the folder

```
import pandas as pd
data = pd.DataFrame({'filename':filenames,'label':categories})

data.to_csv("original_fv.csv",index=False)
```

#Create dataframe for test

```
# Path to the train folder
import os
original_test = '/content/fv_test/'
 
filenames = os.listdir(original_test)
categories = []
for filename in filenames:
    category = filename.split('.')[0]
    if category == 'afr':
        categories.append('0')
    else:
        categories.append('1')

import pandas as pd
data_test = pd.DataFrame({'filename':filenames,'label':categories})

data_test.to_csv("original_fv_test.csv",index=False)
```

```
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
from keras.optimizers import SGD
from keras.preprocessing.image import ImageDataGenerator


#data generators objects

trainval_datagen = ImageDataGenerator(rescale=1./255, horizontal_flip=True, validation_split=0.2)

test_datagen = ImageDataGenerator(rescale=1./255)

#flow_from_dataframe()

train_generator = trainval_datagen.flow_from_dataframe(dataframe=data, 
                                                       directory=original_train,
                                                       x_col='filename',
                                                       y_col='label',
                                                       seed=7,
                                                       shuffle=True,
                                                       subset='training',
                                                       class_mode='binary',
                                                       target_size=(150,150))

val_generator = trainval_datagen.flow_from_dataframe(dataframe=data, 
                                                       directory=original_train,
                                                       x_col='filename',
                                                       y_col='label',
                                                       seed=7,
                                                       shuffle=True,
                                                       subset='validation',
                                                       class_mode='binary',
                                                       target_size=(150,150))



test_generator = test_datagen.flow_from_dataframe(dataframe=data_test, 
                                                  directory=original_test,
                                                  x_col='filename',
                                                  y_col='label',
                                                  seed=7,
                                                  shuffle=False,                                                  
                                                  class_mode=None,
                                                  target_size=(150,150))




#Conv2D is equivalent to a dense layer
#syntax is
#Conv2D(units, (image_width, image_height), padding='valid', activation='relu', 
#       input_shape=(image_width, image_height,image_depth))
#
#
#Dense(units, activation='relu', input_shape=(image_width, image_height,image_depth))


model = Sequential()
model.add(Conv2D(512, (150,150), padding='valid', activation='relu', input_shape=(150,150,3)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

```

```
model.fit_generator(generator=train_generator, validation_data=val_generator, epochs=10, verbose=1)
```

```
model.evaluate_generator(generator=val_generator)
```

```
test_generator.reset()
pred = model.predict_generator(test_generator, verbose=1)
```

```
predfinal = []

for i in pred:
  if i>=0.5:
    predfinal.append(1)
  else:
    predfinal.append(0)


labels = (train_generator.class_indices)

labels = dict((v,k) for k,v in labels.items())
predictions = [labels[k] for k in predfinal]




filename = test_generator.filenames

results = pd.DataFrame({"Filenames":filename, "Predictions":predictions})

```

```
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


cm = confusion_matrix(data_test['label'], results['Predictions'])
```

```
sns.heatmap(cm, annot=True)
plt.show()
```
